# <center>Univariate Synthetic Dataset Generator</center>

```python
import os
import random

import numpy as np
import matplotlib.pyplot as plt
```

## Hyperparameters

```python
random.seed(0)
np.random.seed(0)
```

```python
LEN_ENC = 96
LEN_LABEL = 48
LEN_PRED = 192

DATADIR = '/usr2/home/yongyiw/data/synth'
```

```python
t = np.linspace(-1024, 1024, num=8192)
```

## No Trend

### `sinx`

```python
x = 10 * np.sin(t) + np.random.normal(size=t.shape)
np.save(os.path.join(DATADIR, 'sinx.npy'), x.reshape(-1, 1))

x_plt = x[:LEN_ENC + LEN_PRED]
plt.plot(t[:LEN_ENC + LEN_PRED], x_plt)
plt.vlines(t[LEN_ENC], ymin=min(x_plt), ymax=max(x_plt))
plt.show()
```

### `sinx_sin2x_sin4x`

```python
x = 10 * (np.sin(t) + np.sin(2 * t) + np.sin(4 * t)) + np.random.normal(size=t.shape)
np.save(os.path.join(DATADIR, 'sinx_sin2x_sin4x.npy'), x.reshape(-1, 1))

x_plt = x[:LEN_ENC + LEN_PRED]
plt.plot(t[:LEN_ENC + LEN_PRED], x_plt)
plt.vlines(t[LEN_ENC], ymin=min(x_plt), ymax=max(x_plt))
plt.show()
```

### `xsinx`

```python
a = np.tile(np.exp(np.linspace(0, 4, 512)), len(t) // 512)
x = a * (10 * np.sin(t) + np.random.normal(size=t.shape))
np.save(os.path.join(DATADIR, 'xsinx.npy'), x.reshape(-1, 1))

x_plt = x[:LEN_ENC + LEN_PRED]
plt.plot(t[:LEN_ENC + LEN_PRED], x_plt)
plt.vlines(t[LEN_ENC], ymin=min(x_plt), ymax=max(x_plt))
plt.show()
```

### `sinx_c`

```python
x = 10 * np.sin(t) + np.tile(np.repeat([-30, 30], 32), len(t) // 64) + np.random.normal(size=t.shape)
np.save(os.path.join(DATADIR, 'sinx_c.npy'), x.reshape(-1, 1))

x_plt = x[:LEN_ENC + LEN_PRED]
plt.plot(t[:LEN_ENC + LEN_PRED], x_plt)
plt.vlines(t[LEN_ENC], ymin=min(x_plt), ymax=max(x_plt))
plt.show()
```

## Trend

### `x`

```python
x = t + np.random.normal(size=t.shape)
np.save(os.path.join(DATADIR, 'x.npy'), x.reshape(-1, 1))

x_plt = x
plt.plot(t, x_plt)
plt.show()
```

### `sinx_x`

```python
x = t + 10 * np.sin(t) + np.random.normal(size=t.shape)
np.save(os.path.join(DATADIR, 'sinx_x.npy'), x.reshape(-1, 1))

x_plt = x
plt.plot(t, x_plt)
plt.show()
```

### `sinx_x2_sym`

```python
x = (t/50) ** 2 + 10 * np.sin(t) + np.random.normal(size=t.shape)
np.save(os.path.join(DATADIR, 'sinx_x2_sym.npy'), x.reshape(-1, 1))

x_plt = x
plt.plot(t, x_plt)
plt.show()
```

### `sinx_x2_asym`

```python
x = ((t - np.min(t))/30) ** 2 + 10 * np.sin(t) + np.random.normal(size=t.shape)
np.save(os.path.join(DATADIR, 'sinx_x2_asym.npy'), x.reshape(-1, 1))

x_plt = x
plt.plot(t, x_plt)
plt.show()
```

### `sinx_sqrtx`

```python
x = 20 * np.sqrt(t - np.min(t)) + 10 * np.sin(t) + np.random.normal(size=t.shape)
np.save(os.path.join(DATADIR, 'sinx_sqrtx.npy'), x.reshape(-1, 1))

x_plt = x
plt.plot(t, x_plt)
plt.show()
```
